```python
#categorical columns: Crop_Type, Variety, Soil_Type, Irrigation_Applied
categorical_cols = df[["Crop_Type", "Variety", "Soil_Type"]]
```

```python
from sklearn.preprocessing import OneHotEncoder

#initialize the encoder
encoder = OneHotEncoder(sparse_output=False , handle_unknown='ignore')
encoded_array = encoder.fit_transform(categorical_cols)
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out())
encoded_df
```

```python
df = df.drop(columns=["Crop_Type", "Variety", "Soil_Type"])
```

```python
df = pd.concat([df, encoded_df], axis=1)
```

```python
df
```

```python
df.columns
```

```python
#split the data into x and y
x = df.drop(columns=["Yield_kg_per_Ha"])
y = df["Yield_kg_per_Ha"]

from sklearn.model_selection import train_test_split
x_train ,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_preds=model.predict(x_test)
```

```python
#model evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)
r2 = r2_score(y_test, y_preds)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
model.score(x_test, y_test)
```

```python
corr = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score

scores = cross_val_score(model, x, y, cv=5, scoring='r2')
print("Cross-validation scores:", scores)
print("Mean CV score:", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200,300,400,500],
    'max_depth': [10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(x_train, y_train)

print(f"Best Parameters: {grid_search.best_params_}")
print(f"New R2 Score: {grid_search.best_score_}")
```
